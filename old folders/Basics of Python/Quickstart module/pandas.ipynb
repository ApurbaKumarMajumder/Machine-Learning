```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("bulbasur.png")
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
cv2.imshow
```

```python
plt.axis("off")
plt.imshow(rgb_image);
```

```python
# crop image and show bulbasur's face
crop_image = rgb_image[250:550, 150:450, :]
# plt.axis('off')
plt.imshow(crop_image)
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
```

```python
print(total_pixels)
```

```python
# Resize a image
crop_image = cv2.resize(crop_image,(100, 100))
```

```python
plt.imshow(crop_image);
plt.axis("off")
plt.title("Bulbasur" + str(crop_image.shape))
plt.show()
```

## 5. Pandas - Working with Tabular Data

```python
pokemon = {
    "pikachu" : np.array([59, 89, 75, 64, 22]),
    "bulbasaur" : np.array([82, 73, 135, 58, 133])
}
```

```python
pokemon
```

```python
pokemon = {
    "height" : np.random.randint(1, 100, 5),
    "weight" : np.random.randint(50, 150, 5),
    "power" : np.random.randint(200, 250, 5)
}
```

```python
pokemon
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(pokemon)
print(df) # looks like a excel file
```

```python
df.head(n = 3)
```

```python
print(df.columns)
```
