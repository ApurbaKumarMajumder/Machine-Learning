### JSON Files

JavaScript Object Notation abbreviated as JSON is a light-weight data interchange format. It Encode python objects as JSON strings, and decode JSON strings into Python objects.

```python
from IPython import display
display.Image("json.jpeg")
```

* <strong>json.dump(obj, fileObj) :</strong> Serializes obj as a JSON formatted stream to fileObj.
* <strong>json.dumps(obj) :</strong> Serializes obj as JSON formatted string.
* <strong>json.load(JSONfile) :</strong> De-serializes JSONfile to a Python object.
* <strong>json.loads(JSONfile) :</strong> De-serializes JSONfile(type: string) to a Python object.

```python
import json
```

```python
with open("data.json", "r") as file:
    data = json.load(file)
    print(data["name"])
    print(data["marks"])
    print(data["subjects"][0])
```

```python
with open("data.json", "r") as file:
    data = file.read()
    print(data)
    d = json.loads(data)
    print(d)
```
