* In all the programs we wrote till now, we have designed our program around functions i.e, blocks of statements while manipulate data. This is called the <b>procedure-oriented way of programming</b>.

* There is another way of organizing your program which is to combine data and functionality and wrap it inside something called an object. This is called the <b>object oriented programming paradime</b>.

* Most of the time you can use procedual programming, but when writing large programs or have a program that is better suited to this method, you can use <b> object oriented programming </b> techniques.

* Classes and objects are the two main aspects of object oriented programming. A class creates a new type where objects are instances of the class.

####  Class with a method

class methods have only one specific difference from ordinary functions - they must have an extra first name that has to be added to the beginning of the parameter list, but you do not give a value for this parameter when you call the method, Python will provide it. This particular variable refers to the object itself, and by convention, it is given the name self. 

```python
class Person:
    def say_hi(self):
        self.name = "Apurba"
        print(f"Hello, how are you, {self.name}?")
P = Person()
P.say_hi()
```

#### The \_\_init\_\_ method

```python
class Person:
    # name = "Jatin"
    # not a constructor
    def __init__(self, name):
        self.name = name
        print("New Object Created")
        print()
    def say_hi(self):
        # self.name = "Apurba"
        print(f"Hello, how are you, {self.name}?")

P = Person("abhishek")
P.say_hi()
```

#### Dunders or Magic Functions

```python
class Car:
    def __init__(self, model, mileage):
        self.model = model
        self.mileage = mileage

    def __str__(self):
        return "{} {}".format(self.model, self.mileage)
    
    def __repr__(self):
        return "{}".format(self.model)

    def __eq__(self, other):
        return self.mileage == other.mileage
    
    def __add__(self, other):
        return self.mileage + other.mileage
```

```python
c1 = Car('a', 2)
c2 = Car('b', 2)
```

```python
c1 + c2
```

```python
c1 == c2
```

```python
class Ostream:
    def __lshift__(self, other):
        print(other, end = '')
        return self

cout = Ostream()
```

```python
cout << "apurba" <<" kumar"
```
